<H2>단순 선형 회귀</h2>
1.회귀식 : lm() <br>
2.결과도출 : coef(), fitted(), residuals(), confint(), deviance()<br>
3.회귀 예측과 신뢰구간 : predict()<br>
4.모델 비교 : anova()<br>
5.차트 평가 : plot(lm())
<H2>Summary 조회</h2>
Call : 요청 함수<br>
Residual : 4 분위수<br>
Coef : 변수간의 상관계수, 설명변수, 편차 등 (*이 많으면 상관성이 높은 것임)<br>
종합 : R-squared (표준편차 제곱), F (설명변수), p-value(유의변수 0.05 이하 유의하다고 판단)<br>

lm( )함수 파라미터 <br>
formula : 적용할 모델의 내용으로 보기를 들자면 x ~ y<br>
data : 적용할 모델의 변수를 포함하는 선택적 데이터프레임, 리스트, 또는 환경<br>
subset : fitting process에 사용될 관측치의 서브 세트를 지정하는 optional vector<br>
weights : fitting process에 사용될 가중치의 optional vector<br>


lm(formula, data, subset, weights, na.action, method = "qr", model = TRUE, x = FALSE, y = FALSE, qr = TRUE, singular.ok = TRUE, contrasts = NULL, offset, ...)

```r
RgFit <- lm(Y~X, data=R1)
summary(RgFit)
```

model coefficients  Y = 0.62 * X + 130

```r
coefficients(RgFit) 
```

95% 신뢰구간을 구함
식 : (X1 - X2) / 2
예 : (0.8 - 0.4) / 2 = 0.2 
※ 이 모델은 신뢰구간 95%에서 +/- 0.2

```r
confint(RgFit, level=0.95) # 
fitted(RgFit) # 
```

```r
residuals(RgFit) # ????
deviance(RgFit)
anova(RgFit) # anova table 
vcov(RgFit) # covariance matrix for model parameters 
influence(RgFit) # regression diagnostics
```

```r
par("mar")
par(mar=c(1,1,1,1))
layout(matrix(c(1,2,3,4),2,2)) # optional 4 graphs/page 
plot(RgFit)
plot(RgFit, which=c(4,6))
plot(Y~X, data=R1) 
abline(coef(RgFit))
```

```r
yt.pd=read.csv("power_demand.csv")
fit0=HoltWinters(yt.pd, gamma=F, beta=F)
fit1=HoltWinters(yt.pd, gamma=F, beta=F, alpha=0.1)
fit2=HoltWinters(yt.pd, gamma=F, beta=F, alpha=0.2)
fit3=HoltWinters(yt.pd, gamma=F, beta=F, alpha=0.3)
fit0$SSE;fit1$SSE;fit2$SSE;fit3$SSE
fit9=HoltWinters(yt.pd, gamma=F, beta=F, alpha=0.05)
fit0
predict(fit0, n.ahead=7)
fitted(fit0)[,1]
```

```r
ds0=read.csv("population_vital.csv")
yt.ma=ts(ds0[13:192,4], frequency = 12, start = c(2001,1))
fit.ma=HoltWinters(yt.ma, gamma = F, beta = F, alpha = 0.3)
fit.ma;fit.ma$SSE
predict(fit.ma, n.ahead = 12)
plot(yt.ma, main="Marriage(2001.1.~2015.12)", ylab="(gun)")
lines(predict(fit.ma, n.ahead = 6), col = "blue")
      
```

```r
pv.ma=ts(ds0[13:192,4], frequency = 12, start = c(2001,1))
acf(pv.ma)
Box.test(pv.ma)
Box.test(pv.ma, type="Ljung-Box")
pacf(pv.ma)
```
